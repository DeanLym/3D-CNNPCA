```python
import numpy as np
import h5py
import pickle
import os
```

```python
import sys
sys.path.append('../src/')
```

```python
# Import utils function
from utils import load_model, dot_dict
```

```python
data_dir = './data'
```

```python
args = {}
args['m_petrel'] = os.path.join(data_dir, 'm_petrel_train3000_case1.h5')  # Petrel training models
args = dot_dict(args)
```

```python
# Load m_petrel
m_petrel = load_model(args.m_petrel).astype(np.float32)
m_petrel.shape
```

```python
# Define well locations
well_loc = {}
well_loc['I1'] = (12, 12)
well_loc['I2'] = (28, 3)
well_loc['I3'] = (4, 26)
well_loc['P1'] = (48, 48)
well_loc['P2'] = (58, 31)
well_loc['P3'] = (32, 57)
```

```python
well_hd_all = {}
nz = 40
for wn, (ix, iy) in well_loc.items():
    well_hd_all[wn] = []
    ix -= 1
    iy -= 1
    for iz in range(nz):
        well_hd_all[wn].append((ix, iy, iz, m_petrel[0, iz, iy, ix, 0]))
# Convert to np.array
for wn, hd in well_hd_all.items():
    well_hd_all[wn] = np.array(hd)
```

```python
with open(os.path.join(data_dir, 'hard_data_case2.pickle'), 'wb') as fid:
    pickle.dump(well_hd_all, fid)
```
